```python
# кол-во использования каждой из буквы во всех именах 
import json
from collections import Counter


symbols = []
arr_numb = []
arr_symb = []


count_symbols = {}

for i in names_all['Name']:
    for s in i:
        symbols.append(s.lower())
#         if(s.lower() in cons_word):
#             all_count += names_all['Count']
            
        
c = Counter(symbols)
z = c.most_common()

for v in z:
    arr_numb.append(v[1])
    arr_symb.append(ord(v[0]))

count_symbols['Symbol'] = arr_symb
count_symbols['Number'] = arr_numb

count_symbols
```

```python
df = DataFrame(data=count_symbols)
df
```

```python
df.plot.scatter(x='Symbol', y='Number')
```

```python
names_all_years = {}

cons_word = ["q", "w", "r", "t", "y", "p", "s", "d", "f", "g","h", "j", "k", "l", "z", "x", "c", "v", "b", "n", "m"]

names_by_year = {}
for year in range(1900, 1901):
    names_by_year[year] = pd.read_csv(
        '/Users/Anton/Desktop/python/net/17/names/yob{}.txt'.format(year),
        names=['Name', 'Gender', 'Count'])
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])

n = 0
for i in names_all['Name']:
    z = 0
    if(i in names_all_years):
        names_all_years['{}'.format(i)][0] += names_all['Count'][n]
    else:
        for k in i:
            if(k in cons_word):
                z += 1     
        
        nam = []
        nam.append(names_all['Count'][n])
        nam.append(z)
        names_all_years['{}'.format(i)] = nam

    
    n += 1
names_all_years
```

```python
names_for_pd = []
for i in names_all_years:
    names_for_pd.append({'count': names_all_years[i][0], 'symbols': names_all_years[i][1]})
names_for_pd
```
